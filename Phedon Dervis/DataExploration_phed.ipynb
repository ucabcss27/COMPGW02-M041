### Average bid and payprice per clicked impressions

```python
print('Training Set:')
print('Average bid price of impressions:', round(sum(df_train[df_train['click']==1]['bidprice'])/df_train[df_train['click']==1].shape[0],2))
print('Average pay price of impressions:', round(sum(df_train[df_train['click']==1]['payprice'])/df_train[df_train['click']==1].shape[0],2))
```

### CPM (Cost per Mille)

```python
# Get all advertiser IDs
advertisers = df_train['advertiser'].unique()

#df[df['advertiser']==advertisers[0]]['payprice'].shape[0]
print('Training Set:')
print('')
for a in advertisers:
    
    total_price_paid = sum(df_train[df_train['advertiser']==a]['payprice'])
    total_num_imp = df_train[df_train['advertiser']==a].shape[0]
    cpm = total_price_paid/total_num_imp*1000
    
    print("Advertiser:", a)
    print("CPM :", round(cpm,2))
    print('-----------------------------')
    
```

```python
# Get all advertiser IDs
advertisers_val = df_val['advertiser'].unique()

#df[df['advertiser']==advertisers[0]]['payprice'].shape[0]
print('Validation Set:')
print(' ')
for a in advertisers_val:
    
    total_price_paid = sum(df_val[df_val['advertiser']==a]['payprice'])
    total_num_imp = df_val[df_val['advertiser']==a].shape[0]
    cpm = total_price_paid/total_num_imp*1000
    
    print("Advertiser:", a)
    print("CPM :", round(cpm,2))
    print('-----------------------------')
    
```

## CPC and eCPM

```python
print('Training Set:')
print('')
print('-----------------')
for a in advertisers:
    
    
    num_clicks = df_train[(df_train['advertiser']==a) & (df_train['click']==1)].shape[0]
    total_cost = df_train[(df_train['click'] == 1) & (df_train['advertiser']==a)]['payprice']
    cpc = round(sum(total_cost)/total_cost.shape[0],2)
    
    num_blocks_mille = (df_train[(df_train['advertiser']==a) & (df_train['payprice']>0)].shape[0])/1000
    ecpm = sum(total_cost)/num_blocks_mille
        
    print("Advertiser:", a)
    print("CPC :", round(cpc,2))
    print("eCPM :", round(ecpm,2))
    print('-----------------------------')
    
    #print(ecpc)

```

```python
print('Validation Set:')
print('')
print('-----------------')
for a in advertisers_val:
    
    
    num_clicks = df_val[(df_val['advertiser']==a) & (df_val['click']==1)].shape[0]
    total_cost = df_val[(df_val['click'] == 1) & (df_val['advertiser']==a)]['payprice']
    cpc = round(sum(total_cost)/total_cost.shape[0],2)
    
    num_blocks_mille = (df_val[(df_val['advertiser']==a) & (df_val['payprice']>0)].shape[0])/1000
    ecpm = sum(total_cost)/num_blocks_mille
      
    print("Advertiser:", a)
    print("CPC :", round(cpc,2))
    print("eCPM :", round(ecpm,2))
    print('-----------------------------')
    
    #print(ecpc)

```

## CTR

```python
overall_clicks = df_train[(df_train['click']==1)].shape[0]
total_impressions = df_train.shape[0]
overall_ctr = overall_clicks/total_impressions
overall_ctr
```

```python
for a in advertisers:
    
    num_clicks_adv = df_train[(df_train['advertiser']==a) & (df_train['click']==1)]
    total_imp_adv = df_train[(df_train['advertiser']==a)]
    
    adv_ctr = num_clicks_adv.shape[0]/total_imp_adv.shape[0]
    
    print("Advertiser:", a)
    print("CTR :", round(adv_ctr,6))
    print('-----------------------------')
```

```python
adv1 = df_train[(df_train['advertiser']==2997)]
adv2 = df_train[(df_train['advertiser']==2259)]

avg_payprice1 = adv1.groupby('weekday').payprice.mean()
avg_payprice2 = adv2.groupby('weekday').payprice.mean()

first, = plt.plot(avg_payprice1, color='teal', marker ='o',label='2997')
second, = plt.plot(avg_payprice2, color='orange', marker ='o',label='2259')
plt.xlabel('Day of the week')
plt.ylabel('Average payprice')
plt.title('Average payprice per day')
plt.legend(loc='upper right')
plt.show()
```
