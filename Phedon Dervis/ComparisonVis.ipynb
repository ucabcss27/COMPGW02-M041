```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic strategy comparison

```python
labels = ['constant','randomP','randomIP']
strategies = [0,1,2]
clicks = [15, 15, 55]
clicks2 = [134, 125, 129]

plt.plot(strategies, clicks, 'b--o',label ='val')
plt.xticks(strategies,labels)
plt.title('Clicks by Strategy')
plt.plot(strategies, clicks2, 'r--o',label='train')
plt.xticks(strategies,labels)
plt.legend(loc='lower right')
plt.show()
```

## Advanced model comparison

```python
labels = ['NaiveB','UNaibeB','RandomF','URandomF']
strategies = [0,1,2,4]
clicks = [192, 172, 22, 94]

plt.plot(strategies, clicks, 'b--o')
plt.xticks(strategies,labels)
plt.title('Clicks by Strategy')
plt.ylabel('Clicks')
plt.xticks(strategies,labels)
plt.show()
```
